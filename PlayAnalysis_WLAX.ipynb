```python
import pandas as pd
import numpy as np
```

```python
def playdata_simplify(x):
    playdata = pd.read_excel(x)
    playdata['lead1'] = playdata['X1'].shift(-1)
    playdata['lead2'] = playdata['X1'].shift(-2)
    for i in range(len(playdata)):
        if pd.isna(playdata['X1'][i]):
            playdata['X1'][i] = playdata['lead1'][i]  
    for i in range(len(playdata)):
        if pd.isna(playdata['X1'][i]):
            playdata['X1'][i] = playdata['lead2'][i]  
    for i in range(len(playdata)):
        if pd.isna(playdata['X2'][i]):
            playdata['X2'][i] = playdata['X4'][i]
    playdata['Control'] = np.nan
    teama = playdata['X2'][8]
    teamb = playdata['X4'][8]
    for i in range(len(playdata)):
        if pd.isna(playdata['X4'][i]):
            playdata['Control'][i] = teama
        else:
            playdata['Control'][i] = teamb
    return playdata[['X1','X2','Control']]
```

```python
playdata_simplify('PlayByPlay1.xlsx')
```
